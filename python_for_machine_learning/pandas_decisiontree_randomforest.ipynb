```python
#clean data
df.loc[(df['Years Experience'].isnull()) |
              (df['Employed?'].isnull()) |
              (df['Previous employers'].isnull()) |
              (df['Level of Education'].isnull()) |
              (df['Top-tier school'].isnull()) |
              (df['Interned'].isnull()) |
              (df['Hired'].isnull())] 
```

```python
df.dropna(inplace=True)
df.describe()
```

```python
df
```

```python
d = {'Y': 1, 'N': 0}
df['Hired'] = df['Hired'].map(d)
df['Employed?'] = df['Employed?'].map(d)
df['Top-tier school'] = df['Top-tier school'].map(d)
df['Interned'] = df['Interned'].map(d)

e = {'BS': 0, 'MS': 1, 'PhD': 2}
df['Level of Education'] = df['Level of Education'].map(e)
df.head()
```

```python
all_features = df[['Years Experience', 'Employed?', 'Previous employers',
       'Level of Education', 'Top-tier school', 'Interned']].values

all_classes = df[['Hired']].values

feature_names = ['Years Experience', 'Employed?', 'Previous employers',
       'Level of Education', 'Top-tier school', 'Interned']
```

```python
from sklearn import preprocessing

scaler = preprocessing.StandardScaler()
all_features_scaled = scaler.fit_transform(all_features)
all_features_scaled
```

```python
from sklearn.tree import DecisionTreeClassifier

clf= DecisionTreeClassifier(random_state=1)

# Train the classifier on the training set
clf.fit(all_features, all_classes)
```

```python
print (clf.predict([[10, 1, 4, 0, 0, 0]]))
```

```python
from sklearn.ensemble import RandomForestClassifier

rdf= RandomForestClassifier(n_estimators=10)

# Train the classifier on the training set
rdf.fit(all_features, all_classes)
```

```python
print (rdf.predict([[10, 1, 4, 0, 0, 0]]))
```
