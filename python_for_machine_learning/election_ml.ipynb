```python
import pandas as pd

feature_names =  ['party','handicapped-infants', 'water-project-cost-sharing', 
                    'adoption-of-the-budget-resolution', 'physician-fee-freeze',
                    'el-salvador-aid', 'religious-groups-in-schools',
                    'anti-satellite-test-ban', 'aid-to-nicaraguan-contras',
                    'mx-missle', 'immigration', 'synfuels-corporation-cutback',
                    'education-spending', 'superfund-right-to-sue', 'crime',
                    'duty-free-exports', 'export-administration-act-south-africa']

voting_data = pd.read_csv('C:/Users/sd16053/Desktop/MLCourse/house-votes-84.data.txt', na_values=['?'], 
                          names = feature_names)
voting_data.head()
```

```python
voting_data.dropna(inplace = True)
voting_data.head()
```

```python
voting_data.replace(('y','n'),(1,0),inplace=True)
voting_data.replace(('republican','democrat'),(0,1),inplace=True)
```

```python
voting_data.head()
```

```python
all_features = voting_data[feature_names].drop('party', axis=1).values
all_classes = voting_data[['party']].values
```

```python
import numpy
from sklearn.model_selection import train_test_split

numpy.random.seed(1234)

(training_inputs,
 testing_inputs,
 training_classes,
 testing_classes) = train_test_split(all_features, all_classes, train_size=0.75, random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc= DecisionTreeClassifier(random_state=1)

# Train the classifier on the training set
dtc.fit(training_inputs, training_classes)
```

```python
dtc.score(testing_inputs,testing_classes)
```

```python
from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(dtc, all_features, all_classes, cv=10)
cv_scores.mean()
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc= RandomForestClassifier(random_state=1)

# Train the classifier on the training set
rfc.fit(training_inputs, training_classes)
```

```python
rfc.score(testing_inputs, testing_classes)
```

```python
cv_scores_2 = cross_val_score(rfc, all_features, all_classes, cv=10)
cv_scores_2.mean()
```

```python
#SVM
from sklearn import svm

C = 1.0
svc = svm.SVC(kernel='linear', C=C)
```

```python
cv_scores_3 = cross_val_score(svc,all_features, all_classes, cv=10)
cv_scores_3.mean()
```

```python
#KNN
from sklearn import neighbors

knn = neighbors.KNeighborsClassifier(n_neighbors=10)
cv_scores_4 = cross_val_score(knn, all_features, all_classes, cv=10)

cv_scores_4.mean()
```

```python
#KNN when k is 1 to 50
for n in range(1, 50):
    clf = neighbors.KNeighborsClassifier(n_neighbors=n)
    cv_scores = cross_val_score(clf, all_features, all_classes, cv=10)
    print (n, cv_scores.mean())
```

```python
#Naive Bayes
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
cv_scores_5 = cross_val_score(nb, all_features, all_classes, cv=10)

cv_scores_5.mean()
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression

lo_rgs = LogisticRegression()
cv_scores_6 = cross_val_score(lo_rgs, all_features, all_classes, cv=10)
cv_scores_6.mean()
```
